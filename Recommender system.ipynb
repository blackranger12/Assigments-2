# converting into pivot table 

```python
user_book_df=data.pivot_table(index='user_id',columns='book_title',values='book_rating').reset_index(drop=True)
user_book_df
```

```python
user_book_df.index=data.user_id.unique()
```

```python
user_book_df
```

# filling null(NaN values)

```python
user_book_df.fillna(0,inplace=True)
```

```python
user_book_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

# using cosine simllarity 

```python
user_sim=1-pairwise_distances(user_book_df.values,metric="cosine")
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index=data.user_id.unique()
user_sim_df.columns=data.user_id.unique()
```

```python
user_sim_df
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```
